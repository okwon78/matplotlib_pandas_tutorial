# 2. Legends, Titles and Labels
차트를 그리기 위한 기본적인 Label, Title, Legends 표시 방법을 알아본다.

```python
x = np.arange(start=5, stop=10)
y = np.random.randint(low=10, high=20, size=len(x))

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set(xlabel='x label', ylabel='y label',
       title='About as simple as it gets, folks')

fig.savefig("./img/plot.png")
plt.show()
```

```python
x = np.arange(10)
y1 = np.random.randint(low=1, high=10, size=len(x))
y2 = np.random.randint(low=5, high=15, size=len(x))

fig, ax = plt.subplots()

ax.plot(x, y1, label='legend first Line')
ax.plot(x, y2, label='legend second Line')

# 한번에 설정하는 방법, 추후에 별도 설정
ax.set(xlabel='x label', 
       ylabel='y label',
       title='About as simple as it gets, folks')

ax.legend()

plt.show()
```

# 3. Bar Charts and Histograms


## Bar Charts

Bar 차트는 x, y 값에 대한 막대 그래프를 표시한다. x는 y의 shape과 맞아야 한다.

```python
x = np.array(range(1, 6))
y1 = np.random.randint(low=0, high=100, size=len(x))
y2 = np.random.randint(low=0, high=100, size=len(x))

fig, ax = plt.subplots()
ax.bar(x, y1, label='Bar1', color='blue')
plt.show()

fig, ax = plt.subplots()
ax.bar(x, y2, label='Bar2', color='red')
plt.show()

bar_width = 0.2
fig, ax = plt.subplots()
ax.bar(x+bar_width, y1, label='Bar1', width=bar_width*2, color='c')
ax.bar(x-bar_width, y2, label='Bar2', width=bar_width*2, color='k')

ax.set(xlabel='x label', 
       ylabel='y label',
       title='About as simple as it gets, folks')

ax.legend()
plt.show()
```

```python
# 여러게의 차트를 하나의 이미지로 생성
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.bar(x, y1, label='Bar1', color='blue')

ax2.bar(x, y2, label='Bar2', color='red')

bar_width = 0.4

ax3.bar(x+bar_width, y1, label='Bar1', width=bar_width, color='c')
ax3.bar(x, y2, label='Bar2', width=bar_width, color='k')

# x label
ax3.set_xlabel('Group')
# y label
ax3.set_ylabel('Scores')
# title
ax3.set_title('Scores by group and gender')
# x측 기준점 변경 (위에 방법과 비교)
ax3.set_xticks(x + bar_width / 2)
# x label
ax3.set_xticklabels(('A', 'B', 'C', 'D', 'E'))

ax3.legend()

plt.show()
fig.savefig('./img/barchart.png')
```

## Histograms
Histogram은 Bar차트와는 다르게 x값과 y 값에 대한 단순 표시가 아닌 bins내에 해당 값의 빈도 수를 표시하는 차트이다. 즉 distribution이다.

```python
# 0~10의 값을 랜덤으로 50개 생성한다.
values = np.random.randint(low=0, high=100, size=100)

start = min(values)
stop = max(values)
step = 5

boxes = range(start, stop, step)

fig, ax = plt.subplots()
ax.hist(x=values, bins=boxes, histtype='bar', rwidth=0.8)

ax.set_xlabel('value')
ax.set_ylabel('the number of values')
ax.set_title('histogram')

plt.show()
```

```python
# loc = mean
# scale = standard diviation
values = np.random.normal(loc=20, scale=1, size=10000)

start = int(min(values))
stop = int(max(values))
step = 0.5

bins = np.arange(start, stop, 0.5)

fig, ax = plt.subplots()
# rwidth는 각 bin의 
ax.hist(x=values, bins=bins, rwidth=0.8)
plt.show()
```

# 4. Scatter Plots

Scatter Plot은 일반적으로 2차원 또는 3차원 공간에서의 변수간의 관계를 살펴보기 위해서 많이 사용한다. 예를 들면 x값이 시계열과 같은 값이라면 Line을 사용하는 것이 좋다. 하지만 x간에 관련이 없다면 scatter plot을 이용해야 한다.

```python
y1 = np.random.uniform(low=25, high=75, size=100)
y2 = np.random.normal(loc=50, scale=3.0, size=100)
x = np.arange(start=0, stop=100)

fig, ax = plt.subplots()
ax.scatter(x, y1, label='uniform', color='b', marker="*", s=50)
ax.scatter(x, y2, label='normal', color='r', marker="x", s=50)

ax.set_xlabel('x feature')
ax.set_ylabel('y feature')
ax.set_title('scatter plot')
ax.legend()
plt.show()


# 잘 못된 사용 예를 보여주기 위해 표시함
# x axie 간의 관계가 있을 때만 라인 그래프가 적당하다.
plt.plot(x, y1, label='skitscat', color='b')
plt.show()
```

# 4. Stack Plots

시간변화에 따른 각 항목의 비율을 표시하기 위해 많이 사용되는 그래프

```python
days = np.arange(start=1, stop=6, step=1)

sleeping = np.random.randint(low=5, high=10, size=len(days))
eating = np.random.randint(low=1, high=4, size=len(days))
working = np.random.randint(low=1, high=10, size=len(days))
playing = np.random.randint(low=1, high=4, size=len(days))

labels = ['sleeping', 'eating', 'working', 'playing']

fig, ax = plt.subplots()

#ax.stackplot(days, sleeping, eating, working, playing, labels=labels)
ax.stackplot(days, [sleeping, eating, working, playing], labels=labels)
ax.legend(loc='upper left')
plt.show()
```
