# 2. Legends, Titles and Labels
차트를 그리기 위한 기본적인 Label, Title, Legends 표시 방법을 알아본다.

```python
x = sort([random.randint(1, 10) for _ in range(10)])
y = sort([random.randint(10, 20) for _ in range(10)])

plt.plot(x, y)
plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph\nCheck it out')
```

```python
x1 = sort([random.randint(1, 10) for _ in range(10)])
y1 = sort([random.randint(10, 20) for _ in range(10)])
x2 = sort([random.randint(1, 10) for _ in range(10)])
y2 = sort([random.randint(10, 15) for _ in range(10)])

plt.plot(x1, y1, label='First Line')
plt.plot(x2, y2, label='Second Line')

plt.xlabel('Plot Number')
plt.ylabel('Important var')

plt.title('Interesting Graph\nCheck it out')
plt.legend()
```

# 3. Bar Charts and Histograms


## Bar Charts

Bar 차트는 x, y 값에 대한 막대 그래프를 표시한다.

```python
x1 = [random.randint(10, 20) for _ in range(10)]
y1 = [random.randint(10, 20) for _ in range(10)]
x2 = [random.randint(10, 20) for _ in range(10)]
y2 = [random.randint(10, 20) for _ in range(10)]

plt.bar(x1, y1, label='Bar1', color='c')
plt.bar(x2, y2, label='Bar2', color='r')

plt.xlabel('x label')
plt.ylabel('y label')

plt.legend()
```

## Histograms
Histogram은 Bar차트와는 다르게 x값과 y 값에 대한 단순 표시가 아닌 bins내에 해당 값의 빈도 수를 표시하는 차트이다.

```python
# 0~10의 값을 랜덤으로 1000개 생성한다.
values = [random.randint(0, 10) for _ in range(1000)]
# 0~10의 범위를 설정한다.
ids = [x for x in range(11)]

plt.hist(x=values, bins=ids, histtype='bar', rwidth=0.8)

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('histogram')

plt.legend()
```
