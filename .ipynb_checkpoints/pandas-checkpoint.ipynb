# Pandas tutorial


```python
import pandas as pd
import datetime
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime.now()

df = web.DataReader("XOM", "morningstar", start, end)
```

데이터를 살표 보기 위해서는 print를 사용할 수 있지만, 일반적으로 데이터가 너무 길면 중간 부분이 생략되어 보이지 않는다.
그래서 df.head()를 사용해서 상단 부분만 살펴볼 수 있다.

```python
print(df)
```

```python
print(df.tail(2))
```

matplotlib와 함께 살펴볼 것이다. matplotlib 사용은 다음과 같다.

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

plt.plot(df['High'].values, label='High')
plt.legend()

```

다음과 같이 x축을 지정해 줄 수 있지만 왠지 불편하다. 만약 x축 지정 항목을 사용하지 않을 경우 list의 index 가 사용되어 0부터 시작되게 된다.

```python
style.use('ggplot')

web_stats = {
    'Day':[1,2,3,4,5,6],
    'Visitors':[43,34,65,56,29,76],
    'Bounce Rate':[65,67,78,65,45,52]
}

df = pd.DataFrame(web_stats)

plt.plot(df['Day'].values, df['Visitors'].values, label='Visitors')
plt.plot(df['Day'].values, df['Bounce Rate'].values, label='Bounce Rate')

plt.legend()
```

```python
style.use('ggplot')

web_stats = {
    'Day':[1,2,3,4,5,6],
    'Visitors':[43,34,65,56,29,76],
    'Bounce Rate':[65,67,78,65,45,52]
}

df = pd.DataFrame(web_stats)

plt.plot(df['Visitors'].values, label='Visitors')
plt.plot(df['Bounce Rate'].values, label='Bounce Rate')

plt.legend()
```

dataframe에 x축의 index로 사용할 항목을 다음과 같이 지정할 수 있다. 그러면 line number가 사라지게 된다.

```python
print(df)
df.set_index('Day', inplace=True)
print(df)
```

inplace 파라미터를 지정하지 않으면 재할당을 해줘야 한다.

```python
df = pd.DataFrame(web_stats)
df = df.set_index('Day')
print(df)
```
